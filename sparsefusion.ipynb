```python
# read .pt file
import torch
import sys

def read_pt_file(file_path):
    try:
        data = torch.load(file_path, map_location='cpu')
        return data
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return None
    
def main():
    if len(sys.argv) != 2:
        print("Usage: python read_pt.py <file.pt>")
        sys.exit(1)

    file_path = r"C:\Users\super\Downloads\apple_toy.pt"
    data = read_pt_file(file_path)
    
    if data is not None:
        print(f"Data loaded from {file_path}:")
        print(data)
    else:
        print("Failed to load data.")
        
if __name__ == "__main__":
    main()
```
